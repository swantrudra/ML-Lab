```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('admission.csv')
```

```python
x = dataset['GRE Score']
y = dataset['TOEFL Score']
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("GRE Score")
plt.ylabel("TOEFL Score")
plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar Plot")
plt.xlabel("GRE Score")
plt.ylabel("TOEFL Score")
plt.show()
```

```python
plt.boxplot(x)
plt.title("Box Plot")
plt.show()
```

```python
plt.pie(x)
plt.title("Pie Chart")
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
dataset=pd.read_csv('advertising - advertising.csv')
```

```python
#Scatter Plot (TV Score vs. Newspaper)
plt.figure(figsize=(5,5))
sns.scatterplot(x='TV',y='Newspaper',data=dataset)  
plt.title("Scatter Plot: TV Score vs. Newspaper")
plt.show()
```

```python
#Bar Plot(TV Score vs Radio )
plt.figure(figsize=(5,5))
sns.barplot(x='TV',y='Radio',data=dataset)
plt.title("Bar Plot: TV vs. radio")
```

```python
#Box Plot(Newspaper vs Sales)
plt.figure(figsize=(5,5))
sns.boxplot(x='Newspaper',y='Sales',data=dataset)
plt.title("Box Plot: Newspaper vs. Sales")
```

```python
#Pie chart of total spendings
total_spending = dataset[["TV", "Radio", "Newspaper","Sales"]].sum()
plt.figure(figsize=(6,6))
plt.pie(total_spending, labels=total_spending.index, autopct="%1.1f%%", colors=["blue", "red", "green","yellow"])
plt.title("Total Spending Distribution by Advertisement Channel")
plt.show()
```

```python
#Line Chart (Trend of Sales Across Data Points)
plt.figure(figsize=(5,5))
plt.plot(dataset["Sales"], marker="o", color="green", linestyle="dashed")
plt.title("Sales Trend Over Time")
plt.xlabel("Data Point Index")
plt.ylabel("Sales ($1000s)")
plt.show()
```
